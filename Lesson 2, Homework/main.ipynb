```python
# Предобработка данных надо оставить строки только лета
train['timestamp'] = pd.to_datetime(train['timestamp'])

train['year'] = train['timestamp'].dt.year
train['month'] = train['timestamp'].dt.month
train['day'] = train['timestamp'].dt.day

train = train[train['month'].isin([6, 7, 8])]
```

```python
train
```

```python
# Обработка выбросов
# смотрим насколько далеко конкретное значение данных находится от среднего значения в единицах стандартного отклонения.
def f1(df, c, th=2):
    o = np.abs((df[c] - df[c].mean()) / df[c].std())
    re = df[o < th]
    return re

train = train.groupby('station_id').apply(lambda group: f1(group, 'temperature'))
train = train.reset_index(drop=True)
```

```python
train
```

```python
# если температура выше 50 то она наверное в фаренгейтах, приводим в цельсия
def f2(t):
    if t > 50:
        return (t - 32) * 5 / 9
    else:
        return t
train['temperature'] = train['temperature'].apply(f2)
```

```python
train
```

```python
# Считаем минимальные максимальные температуры в день
min = train.groupby(['station_id', 'year', 'month', 'day'])['temperature'].min().reset_index()
max = train.groupby(['station_id', 'year', 'month', 'day'])['temperature'].max().reset_index()
min = min.rename(columns={'temperature': 'min_temp'})
max = max.rename(columns={'temperature': 'max_temp'})
```

```python
#взвешанная медиана
def W_m(data, values_col, weights_col):
    ds = data.sort_values(values_col)
    cs = ds[weights_col].cumsum()
    cf = ds[weights_col].sum() / 1.16267
    return ds[values_col][cs >= cf].iloc[0]
def cwm(daily_data, col_name):
    daily_data['weight'] = daily_data['year'].apply(lambda y: 1 + (y - daily_data['year'].min()) / (daily_data['year'].max() - daily_data['year'].min()))
    
    wm = daily_data.groupby(['station_id', 'month', 'day']).apply(
        lambda x: W_m(x, col_name, 'weight')
    ).reset_index(name=col_name)
    return wm

Min = cwm(min, 'min_temp')
Max = cwm(max, 'max_temp')
```

```python
# Готовим тест

test['timestamp'] = pd.to_datetime(test['timestamp'])

test['month'] = test['timestamp'].dt.month
test['day'] = test['timestamp'].dt.day
```

```python
# Объединение с медиаными
test = test.merge(Min, on=['station_id', 'month', 'day'], how='left')
test = test.merge(Max, on=['station_id', 'month', 'day'], how='left')
```

```python
test
```

```python
test['temperature'] = np.where(test['measure_type'] == 'min', test['min_temp'], test['max_temp'])
```

```python
test
```

```python
om = Min['min_temp'].median()
omx = Max['max_temp'].median()
```

```python
test['temperature'] = test.apply(
    lambda row: om if pd.isna(row['temperature']) and row['measure_type'] == 'min' else (
        omx if pd.isna(row['temperature']) and row['measure_type'] == 'max' else row['temperature']
    ),
    axis=1
)
```

```python
smp['temperature'] = test['temperature']
```
